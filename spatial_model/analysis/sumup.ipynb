# **Flock Model: analysis of parameters**

Spatial model of collective motion

* **Zone of repulsion:**
$$
\large\overrightarrow{d_i} = -\sum\limits_{j \neq i}{\frac{\overrightarrow{c_j} - \overrightarrow{c_i}}{|\overrightarrow{c_j} - \overrightarrow{c_i}|}}
$$

* **Zone of attraction and allignment:**
$$
\large\overrightarrow{d_i} = \sum\limits_{j \neq i}{\frac{\overrightarrow{c_j} - \overrightarrow{c_i}}{|\overrightarrow{c_j} - \overrightarrow{c_i}|}} + \sum\limits_{j}{\frac{\overrightarrow{v_j}}{|\overrightarrow{v_j}|}}
$$

* **Add personal preference:**
$$
\large\overrightarrow{d_i}^{'} =\frac{\hat{d_i}+\omega \hat{g_i}}{|\hat{d_i}+\omega \hat{g_i}|}
$$

* **Positive or negative feedback on $\omega$**

## `beta`

In some simulations the group splits and we want to prevent that. I introduced a the term in order to tune attraction and allignment (`beta`). In this case the model follow the rule:

$$
\large\overrightarrow{d_i} = \sum\limits_{j \neq i}{\frac{\overrightarrow{c_j} - \overrightarrow{c_i}}{|\overrightarrow{c_j} - \overrightarrow{c_i}|}} + \sum\limits_{j}{\frac{\overrightarrow{v_j}}{|\overrightarrow{v_j}|}}
$$

$$
\large\overrightarrow{d_i}^{'} =\frac{\beta\hat{d_i}+\omega \hat{g_i}}{|\beta\hat{d_i}+\omega \hat{g_i}|} = \frac{\hat{d_i}+\frac{\omega}{\beta} \hat{g_i}}{|\hat{d_i}+\frac{\omega}{\beta} \hat{g_i}|}
$$

For the following plot I used `angular_tresh=40deg` and the same initial conditions.

**The parameter `beta` can avoid the group splitting**

But it is important to be sure that it doesn't affect the trajectory and pattern of the simulation without it.

I now want to check if the introduction of `beta` parameter changes the bifurcation point.

```python
fig, ax = plt.subplots(figsize=(8,4))

beta = []
bif_point =[]

#-------------------------------- ANG TRESH 40 --------------------------------

df_cen = pd.read_csv('../output/bif_point/centroid_b1_deg40.txt', sep='\t')
df_tar = pd.read_csv('../output/bif_point/targets_b1_deg40.txt', sep='\t')
num_max = len(df_cen[df_cen['cue_reached']!=-1].replicate.unique())

for i in range(1,16):
    df_cen = pd.read_csv(f'../output/bif_point/centroid_b{i}_deg40.txt', sep='\t')
    reached = df_cen[df_cen['cue_reached']!=-1].replicate.head(num_max+1)
    df_cen = df_cen[df_cen.replicate.isin(reached)]
    beta.append(i)
    bif_point.append(bifurcation_point(df_cen))
    
bif_point[:] = [x / df_tar.loc[df_tar.id==0,:].x for x in bif_point]
plt.scatter(beta[:], bif_point[:], label = 'Angular tresh 40')
plt.xticks(np.arange(0, 16), size=12)
plt.yticks(np.arange(0, 1, 0.1), size=12)
plt.xlabel('Beta', size=18)
plt.ylabel('Normalized bifurcation point', size=18)

bp1 = np.array(bif_point)

#-------------------------------- ANG TRESH 20 --------------------------------

beta = []
bif_point =[]

for i in range(1,16):
    df_cen = pd.read_csv(f'../output/bif_point/centroid_b{i}_deg20.txt', sep='\t')
    reached = df_cen[df_cen['cue_reached']!=-1].replicate.head(num_max)
    df_cen = df_cen[df_cen.replicate.isin(reached)]
    beta.append(i)
    bif_point.append(bifurcation_point(df_cen))
       
bif_point[:] = [x / df_tar.loc[df_tar.id==0,:].x for x in bif_point]
plt.scatter(beta[:], bif_point[:], label = 'Angular tresh 20')

#plt.xticks(np.arange(0, 16), size=12)
plt.yticks(np.arange(0, 1, 0.1), size=12)
plt.xlabel('Beta', size=14)
plt.ylabel('Bifurcation point', size=14)
plt.title('Normalized bifurcation point\n', size=16)
plt.legend(loc=1);
```

## Bifurcation point over angular thershold (Beta=1)

Going back to the `angular_thresh` parameter I now study the **bifurcation point** as a function of the value of the **angular threshold**

```python
fig, ax = plt.subplots(figsize=(8,4))

num_reached = []

for i in range(3,36):
    df_cen = pd.read_csv(f'../output/bif_over_angthresh/centroid_{i}deg.txt', sep='\t')
    num_reached.append(len(df_cen[df_cen['cue_reached']!=-1].replicate.unique()))
#print(min(num_reached))

ang_thresh = []
bif_point =[]
df_tar = pd.read_csv('../output/bif_over_angthresh/targets_5deg.txt', sep='\t')
    
for i in range(3,36):
    df_cen = pd.read_csv(f'../output/bif_over_angthresh/centroid_{i}deg.txt', sep='\t')
    reached = df_cen[df_cen['cue_reached']!=-1].replicate.head(min(num_reached))
    df_cen = df_cen[df_cen.replicate.isin(reached)]
    ang_thresh.append(i)
    bif_point.append(bifurcation_point(df_cen))
    
bif_point[:] = [x / df_tar.loc[df_tar.id==0,:].x for x in bif_point]
plt.scatter(ang_thresh, bif_point, s=20, alpha = 1)
plt.xticks(np.arange(0, 40, 5), size=12)
plt.yticks(np.arange(0, 0.8, 0.1), size=12)
plt.title('Bifurcation point over angular thershold (Beta=1)\n', size=16)
plt.xlabel('Angular Threshold', size=14)
plt.ylabel('Normalized bifurcation point', size=14);
```
