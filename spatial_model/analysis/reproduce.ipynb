# Analysis of centroid's trajectory
## Reproduce Vivek's results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

## 2 targets

```python
# display parameters
text_file = open("../output/reproduce/parameters_2_targets.txt")
file_content = text_file.read()
print(file_content)
text_file.close()

df_cen = pd.read_csv('../output/reproduce/centroid_2_targets.txt', sep='\t')
df_tar = pd.read_csv('../output/reproduce/targets_2_targets.txt', sep='\t')

fig, ax = plt.subplots(figsize=(7, 5))

for i in range(df_cen.replicate.max()+1):
    plt.plot(df_cen.loc[df_cen.replicate==i,:].x, df_cen.loc[df_cen.replicate==i,:].y, c='black', alpha=0.3)

plt.scatter(df_tar.loc[df_tar.id==0,:].x, df_tar.loc[df_tar.id==0,:].y, c='r', s=100)
plt.scatter(df_tar.loc[df_tar.id==1,:].x, df_tar.loc[df_tar.id==1,:].y, c='r', s=100)

# set x and y labels as x and y
plt.xlabel('X (a.u.)', fontsize=16)
plt.ylabel('Y (a.u.)', fontsize=16)

# set tick label size
ax.tick_params(axis='both', which='major', labelsize=14)

# save figure in pdf dpi 100
plt.savefig('../img/2_targets.pdf', dpi=100)
```
